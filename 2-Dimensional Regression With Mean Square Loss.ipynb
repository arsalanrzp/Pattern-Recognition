```python
import numpy as np

np.random.seed(0)
a, b = 2, 3
num_data_points = 50

x_values = np.random.uniform(0, 10, num_data_points)

y_values = a * x_values + b + np.random.normal(0, 1, num_data_points)

data = np.column_stack((x_values, y_values))
```

```python
Sigma_XiXi = np.inner(x_values, x_values)
Sigma_Xi = np.sum(x_values)
Sigma_one = num_data_points
Sigma_XiYi = np.inner(x_values, y_values)
Sigma_Yi = np.sum(y_values)
A = np.matrix([[Sigma_XiXi, Sigma_Xi], [Sigma_Xi, Sigma_one]])
B = np.matrix([[Sigma_XiYi], [Sigma_Yi]])
Response = np.inner(np.linalg.inv(A), B.T)
print(Response)
print(Response.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_values, y_values, color='blue', label='Data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x against y')

x_line = np.linspace(0, 20, 1000)
y_line = a * x_line + b
yhat_line = Response[0,0] * x_line + Response[1,0]
plt.plot(x_line, y_line, color='red', label='Line: y = 3x + 1')
plt.plot(x_line, yhat_line, color='green', label='Line: y = 3x + 1')
plt.title(f'Scatter Plot with Line y = {a}x + {b} and data points')
plt.grid(True)
plt.legend()
plt.show()
```
